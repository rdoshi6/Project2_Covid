```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless = False)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
data_set = soup.find_all('tr')
```

```python
data_set[0:10]
```

```python
trs = []
```

```python
for data in data_set:
    stuff = data.find_all("td")
    #print(stuff)
    trs.append(stuff)
```

```python
del trs[0]
```

```python
dirtstates = []
dirttested = []
dirtpositive = []
```

```python
for i in trs:
    dirtstates.append(i[0])
    dirttested.append(i[1])
    dirtpositive.append(i[2])
```

```python
states = []
for state in dirtstates:
    states.append(state.text)
```

```python
tested = []
for test in dirttested:
    tested.append(test.text)
```

```python
positive = []
for percent in dirtpositive:
    positive.append(percent.text)
```

```python
print(f'''
{len(states)}
{len(tested)}
{len(positive)}''')
```
