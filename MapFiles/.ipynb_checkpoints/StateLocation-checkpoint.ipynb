```python
from bs4 import BeautifulSoup
import pandas as pd
from splinter import Browser
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless = False)
```

```python
url = 'https://inkplant.com/code/state-latitudes-longitudes'
```

```python
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
data_set = soup.find_all("tr")
```

```python
trs = []
for data in data_set:
    stuff = data.find_all("td")
    #print(stuff)
    trs.append(stuff)
```

```python
trs
```

```python
dirtStates = []
dirtLat = []
dirtLng = []
```

```python
for i in trs:
    dirtStates.append(i[0])
    dirtLat.append(i[1])
    dirtLng.append(i[2])
```

```python
States = []
for state in dirtStates:
    States.append(state.text)
```

```python
Lat = []
for lat in dirtLat:
    Lat.append(lat.text)
```

```python
Lng = []
for lng in dirtLng:
    Lng.append(lng.text)
```

```python
df = pd.DataFrame(list(zip(States, Lat, Lng)), 
               columns =['State', 'Lat', 'Lng'])
df
```
