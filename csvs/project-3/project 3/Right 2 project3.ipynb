```python
temp = data3[data3['Date']==max(data3['Date'])]

m = folium.Map(location=[0, 0], titles='cartodbpositron', min_zoom = 1, max_zoom=4, zoom_start=1)
               
for i in range(0, len(temp)):
    folium.Circle(location=[temp.iloc[i]['Lat'], temp.iloc[i]['Long']], color = 'crimson', fill = 'crimson', 
                 tooltip = '<li><bold> Province_State: ' + str(temp.iloc[i]['Province_State'])+
                             '<li><bold> Admin2: ' + str(temp.iloc[i]['Admin2'])+
                               '<li><bold> Deaths: ' + str(temp.iloc[i]['Deaths']),
radius = int(temp.iloc[i]['Deaths'])**0.5).add_to(m)
   
m
```
